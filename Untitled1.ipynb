```python
import pandas as pd
sales_data = pd.read_csv('/content/sample_data/Sales_Data.csv')
sales_data.head()
```

```python
!pip install joblib
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
features = sales_data[['Quantity', 'Unit Price', 'Net Price', 'Discount (%)']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method to Find Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
sales_data['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=sales_data['Cluster'], cmap='viridis')
plt.xlabel('Quantity (scaled)')
plt.ylabel('Unit Price (scaled)')
plt.title('Customer Clusters')
plt.show()
```
